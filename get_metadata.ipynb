```python
import pandas as pd
import numpy as np
import json
from PIL import Image
```

```python
df = pd.read_excel('src/static/Shufu no tomo Metadata.xlsx')
df = df.drop([0, 1])
```

```python
results = []
for i, row in df.iterrows():
    img = Image.open("src/static/thumbs/" + row['file name FRONT'])
    
    append = {
        'src': row['file name FRONT'],
        'src_back': row['file name BACK'],
        'title_japanese': row['card title on front'],
        'title_english': row['card title on front in'],
        'width': img.size[0],
        'height': img.size[1],
        'year': row['date of card']
    }
    subjects = []
    for i in range(1, 6):
        if row[f"subject {i}"] == row[f"subject {i}"]:
            subjects.append(row[f"subject {i}"].strip().lower())
    append['subjects'] = subjects
    results.append(append)
```

```python
with open('./src/static/data.json', 'w') as f:
    json.dump(results, f)
```
